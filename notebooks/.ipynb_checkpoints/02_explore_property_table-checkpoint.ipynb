# Property Table Analysis

This notebook profiles the Airtable **Property** table and optionally cross-checks it against incidences exported in `data/airtable_sample/incidences_sample.csv`.

Goals:
- Understand volume and coverage of key business fields
- Detect nulls, duplicates, and schema inconsistencies
- Validate join feasibility with incidences (record id / unique id)
- Produce conclusions for graph modeling decisions


```python
from pathlib import Path
import pandas as pd
import numpy as np

pd.set_option('display.max_columns', 120)
pd.set_option('display.width', 200)

ROOT = Path.cwd().resolve().parent
AIRTABLE_DIR = ROOT / 'data' / 'airtable_sample'

PROPERTY_CANDIDATES = [
    AIRTABLE_DIR / 'properties_sample.csv',
    AIRTABLE_DIR / 'property_sample.csv',
    AIRTABLE_DIR / 'properties.csv',
    AIRTABLE_DIR / 'Property.csv',
]

INCIDENCES_CSV = AIRTABLE_DIR / 'incidences_sample.csv'

def pick_existing(paths):
    for p in paths:
        if p.exists():
            return p
    return None

PROPERTY_CSV = pick_existing(PROPERTY_CANDIDATES)
print('ROOT:', ROOT)
print('Property CSV:', PROPERTY_CSV)
print('Incidences CSV exists:', INCIDENCES_CSV.exists())
```

```python
if PROPERTY_CSV is None:
    raise FileNotFoundError(
        'Property CSV not found under data/airtable_sample/. '
        'Export Airtable Property table first (e.g., properties_sample.csv).'
    )

props = pd.read_csv(PROPERTY_CSV)
props.head(3)
```

```python
print('Rows:', len(props))
print('Columns:', len(props.columns))
print('Column names:')
for c in props.columns:
    print('-', c)
```

## Data quality overview


```python
nulls = pd.DataFrame({
    'column': props.columns,
    'null_count': props.isna().sum().values,
    'null_pct': (props.isna().sum().values / max(len(props), 1) * 100).round(2),
}).sort_values(['null_pct', 'null_count'], ascending=False)
nulls.head(30)
```

```python
# Approx cardinality scan
cardinality = []
for c in props.columns:
    s = props[c]
    cardinality.append({
        'column': c,
        'dtype': str(s.dtype),
        'non_null': int(s.notna().sum()),
        'unique_non_null': int(s.nunique(dropna=True)),
    })

card = pd.DataFrame(cardinality).sort_values('unique_non_null', ascending=False)
card.head(30)
```

## Suggested business fields coverage

Expected candidate fields to inspect (names may differ):
- address
- rent
- type
- insurance
- maintenance and renovations
- property conditions
- property manager
- tenant name
- tenant phone
- tenant e-mail
- record id for joins


```python
keywords = [
    'address', 'rent', 'type', 'insurance', 'maint', 'renov', 'condition',
    'manager', 'tenant', 'phone', 'email', 'e-mail', 'record', 'id',
]

matches = []
for c in props.columns:
    lc = c.lower()
    if any(k in lc for k in keywords):
        matches.append(c)

matches
```

```python
# Coverage table for matched columns
coverage_rows = []
for c in matches:
    s = props[c]
    coverage_rows.append({
        'column': c,
        'non_null': int(s.notna().sum()),
        'non_null_pct': round(float(s.notna().mean() * 100), 2),
        'unique_non_null': int(s.nunique(dropna=True)),
        'example': str(s.dropna().astype(str).head(1).tolist()[0]) if s.dropna().shape[0] else None,
    })

coverage = pd.DataFrame(coverage_rows).sort_values('non_null_pct', ascending=False) if coverage_rows else pd.DataFrame()
coverage
```

## Optional join check with incidences

This section helps estimate whether the Property table can be linked reliably to incidences using IDs.


```python
inc = pd.read_csv(INCIDENCES_CSV) if INCIDENCES_CSV.exists() else None
print('Incidences rows:' , 0 if inc is None else len(inc))

if inc is not None:
    print('Incidences columns with id hints:')
    for c in inc.columns:
        if any(k in c.lower() for k in ['id', 'unique', 'record', 'property']):
            print('-', c)
```

```python
# Heuristic candidate join columns
prop_id_candidates = [c for c in props.columns if any(k in c.lower() for k in ['record', 'id', 'unique'])]
inc_id_candidates = [] if inc is None else [c for c in inc.columns if any(k in c.lower() for k in ['record', 'id', 'unique', 'property'])]

print('Property id candidates:', prop_id_candidates)
print('Incidence id candidates:', inc_id_candidates)
```

```python
# Try simple overlap checks (string-normalized)
if inc is not None and prop_id_candidates and inc_id_candidates:
    rows = []
    for pc in prop_id_candidates:
        pvals = set(props[pc].dropna().astype(str).str.strip())
        for ic in inc_id_candidates:
            ivals = set(inc[ic].dropna().astype(str).str.strip())
            overlap = len(pvals & ivals)
            rows.append({
                'property_col': pc,
                'incidence_col': ic,
                'property_unique': len(pvals),
                'incidence_unique': len(ivals),
                'overlap': overlap,
            })
    overlap_df = pd.DataFrame(rows).sort_values('overlap', ascending=False)
    overlap_df.head(20)
else:
    print('Not enough candidate columns to test join overlap.')
```
